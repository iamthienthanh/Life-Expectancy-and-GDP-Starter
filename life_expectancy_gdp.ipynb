```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['GDP'])
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Life expectancy at birth (years)'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', data=df)
```

```python
df[['Life expectancy at birth (years)', 'GDP']].corr()
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(1,2,1)
sns.lineplot(x='Year', y='Life expectancy at birth (years)', hue='Country', data=df)
plt.title('Life expectancy at birth (years) over time')
plt.subplot(1,2,2)
sns.lineplot(x='Year', y='GDP', hue='Country', data=df)
plt.title('GDP over time')
```
